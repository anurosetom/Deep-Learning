```python
import collections
ftrain = open('E:\\input_data\\training.txt','r',encoding='UTF8')
print(ftrain)
maxlen=0
word_freqs=collections.Counter()
num_recs=0
```

```python
import nltk

for line in ftrain:
    label,sentence=line.strip().split('\t')
    words=nltk.word_tokenize(sentence.lower())
    #sentence were tokenizing into lowercase for avoiding words's case match(there will be lower case&uppercase words)
    #print(words)
    #print(len(words))
    
    if len(words)>maxlen:
        maxlen=len(words)
    for word in words:
        word_freqs[word]+=1
    num_recs+=1
ftrain.close()
```

```python
# Most common 2000 words's & index position will be displayed
#i=0 for first time, so 1st index will be 2

word2index={x[0]:i+2 for i,x in enumerate(word_freqs.most_common(MAX_FEATURES))}
#then we need to include 2 words-PAD & UNK(Unknown) to this .so
word2index['PAD']=0
word2index['UNK']=1
print(word2index)
#THEN PAD,UNK got added to the end of dictionary set
print(len(word2index))
```

```python
ftrain = open('E:\\input_data\\training.txt','r',encoding='UTF8')
i=0
for line in ftrain:
    label,sentence=line.strip().split('\t')
    words=nltk.word_tokenize(sentence.lower())
    seqs=[]
    for word in words:
        if word in word2index:
            seqs.append(word2index[word])
        else:
            seqs.append(word2index['UNK'])
    x[i]=seqs
    y[i]=int(label)
    i+=1
ftrain.close()
    
#initially label is string,which is convert into integer
```
