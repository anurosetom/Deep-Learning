```python
import json
data_file= open('E:\pyth_env\intents.json').read()
#print(data_file)
```

```python
intents=json.loads(data_file)
#to get json objects to data file it use json.loads
print(intents)
```

```python
#convert this test data into response
#need to do natural language processing
#then lemmatizing(root words identifying....)
#for that we need to tokenize 
#then identify sentence like in which class they need to come
#then into documents
#then remove special characters(? and ! here)
#here keys are "intents",so we can metion intents as "key"
#and we loaded these dictionary into "intents" initially and its values as "patterns" in file

import nltk
from nltk.stem import WordNetLemmatizer
lemmatizer=WordNetLemmatizer()
words=[]
classes=[]
documents=[]

ignore_words=['?','!']
for intent in intents['intents']:
    for pattern in intent['patterns']:
        #print(pattern)
        #then tokenize each pattern
        w=nltk.word_tokenize(pattern)
        #print(w)
        #w is a list type value where each words in sentence separated
        #then we need to append to words(an empty list),ie, we use extend()
        words.extend(w)
#print(words)
        #then appending tags to those sentence like is it a greeting type or anything
        documents.append((w,intent['tag']))
#print(documents)
        #then giving that as a response after identifying the request
        #for that we need to append these tags to class
        #to avoid repeating of class ,an if condition given
        if intent['tag'] not in classes:
            classes.append(intent['tag'])
#print(words)#each words including those special characters
#print(documents)#tuple form
print(classes)#type of sentence
print(len(classes))
        
        
```

```python
#to remove special characters from words
for w in words:
    if w not in ignore_words:
        print(w)
```

```python
#list comprehension method
#then change the case to lower case
words=[lemmatizer.lemmatize(w.lower()) for w in words if w not in ignore_words]
print(len(words))
#print(words)
words=sorted(list(set(words)))
#to avoid those repeated words, make it into set(only display unique words)
print(len(words))
#print(words)

classes=sorted(list(set(classes)))
print(classes)
print(len(classes))
```

```python
print(len(documents),"documents")
print(len(classes),"classes",classes)
print(len(words),"unique lemmatized words",words)
```

```python
#to identify the question by user(in which class,that question contained)
#pickle is a format to save them into file format
import pickle
pickle.dump(words, open('words.pkl','wb'))
pickle.dump(classes, open('classes.pkl','wb'))
```

```python
#Now train the data
```

```python
training=[]
output_empty=[0]*len(classes)
#print(output_empty)

for doc in documents:
    bag=[]
    pattern_words=doc[0]
    #print(pattern_words)
    pattern_words= [lemmatizer.lemmatize(w.lower()) for w in pattern_words]
    #print(pattern_words)
    for w in words:
        bag.append(1) if w in pattern_words else bag.append(0)
    #print(len(bag))
    #print(bag)
    output_row=list(output_empty)
    output_row[classes.index(doc[1])]=1
    #print(output_row)
    #output row[4th position]that is 1
    
    #then bag and output row appending
    training.append([bag,output_row])
    #print(training)
    
#then shuffling the documents(because data points are same and are nearby)
import numpy as np
import random
random.shuffle(training)
training=np.array(training)
print(training)
```

```python
train_x=list(training[:,0])
train_y=list(training[:,1])
print("Training data created")
```

```python
from keras.models import Sequential
from keras.layers import Dense,Activation,Dropout
from keras.optimizers import SGD

#CREATING 3 LAYERS
#1.
#2.this layer =>having 64neurons
#3.Output layer=>having 9 classes ,so 9 neurons

model=Sequential()
model.add(Dense(128,input_shape=(len(train_x[0]),),activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(64,activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(len(train_y[0]),activation='softmax'))

#model created now
#Next step is optimize

sgd=SGD(lr=0.01,decay=1e-6,momentum=0.9,nesterov=True)
#1e-6 is 10 to the power of -6
model.compile(loss='categorical_crossentropy',optimizer=sgd,metrics=['accuracy'])
hist=model.fit(np.array(train_x),np.array(train_y),epochs=200,batch_size=5,verbose=1)
model.save('chatbot_model.h5',hist)
print("Model Created")
```

```python
from keras.models import load_model
import json
import pickle
model=load_model('chatbot_model.h5')
intents=json.loads(open('E:\pyth_env\intents.json').read())
#open saved files then
words=pickle.load(open('words.pkl','rb'))
classes=pickle.load(open('classes.pkl','rb'))
print(words)
```

```python
#Creating a function for preprocessing

import numpy as np
import random
import nltk
from nltk.stem import WordNetLemmatizer
from nltk import word_tokenize
lemmatizer=WordNetLemmatizer()

#Function for Tokenize words
def clean_up_sentence(sentence):
    sentence_words=nltk.word_tokenize(sentence)
    #then lemmatizing
    sentence_words=[lemmatizer.lemmatize(word.lower()) for word in sentence_words]
    return sentence_words

#Function for bag of word representation.Inside that it need to pass sentence and words)

def bow(sentence,words,show_details=True):
    sentence_words=clean_up_sentence(sentence)
    #above sentence words contains that return value from clean_up functions end
    bag=[0]*len(words)
    for s in sentence_words:
        for i,w in enumerate(words):
            if w==s:
                bag[i]=1
                if show_details:
                    print("found in bag:%s"%w)
    return (np.array(bag))
 
    
#Next Function is to predict Output

def predict_class(sentence,model):
    #here sentence is the question from user
    #so there what happens is calling sentence,then words(bag of words function,and clean_up_function will be called)
    
    p=bow(sentence,words,show_details=False)
    #here p is bag of word representation
    res=model.predict(np.array([p]))[0]
    #then setting a threshold value
    ERROR_THRESHOLD=0.25
    results=[[i,r] for i,r in enumerate(res) if r>ERROR_THRESHOLD]
    #r is the positon & probability
    results.sort(key=lambda x: x[1], reverse=True)
    return_list=[]
    for r in results:
        return_list.append({"intent":classes[r[0]],"probability":str(r[1])})
        #classes il rth position
    return return_list
    
    
#Next Function is to give response

def getResponse(ints,intents_json):
    #now need to match tag with corresponding response
    tag=ints[0]['intent']
    #ints[0] now contain this dictionary =>{"intent":classes[r[0]],"probability":str(r[1])}
    list_of_intents=intents_json['intents']
    for i in list_of_intents:
        if(i['tag']==tag):
            result=random.choice(i['responses'])
    return result

def chatbot_response(msg):
    ints=predict_class(msg,model)
    #then calling getResponse
    res=getResponse(ints,intents)
    #intents is json file
    return res
```

```python
import tkinter
from tkinter import *

#we need to create a function to get response from user
def send():
    msg=EntryBox.get("1.0",'end-1c').strip()
    #strip removes space
    EntryBox.delete("0.0",END)
    if msg != '':
        ChatLog.config(state=NORMAL)
        ChatLog.insert(END, "You: " + msg + '\n\n')
        ChatLog.config(foreground="#442265", font=("Verdana", 12))
        #taking the predicted response
        res = chatbot_response(msg)
        ChatLog.insert(END, "Bot: " + res + '\n\n')
        
        ChatLog.config(state=DISABLED)
        ChatLog.yview(END)
        
base = Tk()
#below things for chatbot's heading,size,scrollbar size, etc
base.title("hello")
base.geometry("400x500")
base.resizable(width=FALSE, height=FALSE)

ChatLog = Text(base, bd=0, bg="white",width="29", height="5", font="Arial") 
ChatLog.config(state=DISABLED)

scrollbar=Scrollbar(base,command=ChatLog.yview, cursor='heart')
ChatLog['yscrollcommand'] = scrollbar.set

SendButton = Button(base, font=("Verdana",12,'bold'), text="Send", width="12" ,height=5,bd=0, bg="#32de97", activebackground="#3c9d9b",fg='#ffffff',
                    command= send )
EntryBox = Text(base, bd=0, bg="white",width="29", height="5", font="Arial")

scrollbar.place(x=376,y=6, height=386)
ChatLog.place(x=6,y=6, height=386, width=370)
EntryBox.place(x=128, y=401, height=90, width=265)
SendButton.place(x=6, y=401, height=90)

base.mainloop()
```
